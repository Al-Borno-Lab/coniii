```python
import networkx as nx
```

```python
n = 9
adj = ones((n,n))
adj[diag_indices(n)] = 0

G = nx.Graph(adj)
```

```python
fig,ax = plt.subplots(figsize=(6.4,6))
nx.draw_circular(G,ax=ax,node_color=choice(['k','w'],size=n),
                 edge_color=choice(['r','b'],size=36),width=2)
fig.savefig('ising_example.png',frameon=False,transparent=True,bbox_inches='tight',pad_inches=1)
```
